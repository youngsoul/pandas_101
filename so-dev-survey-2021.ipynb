## Apply, ApplyMap, Map, Replace

### Apply

DataFrame or Series

Series:  apply a function to every value in the series and you return the modified value

DataFrame: apply a function to every row or column 

```python
def make_lower(val):
    return val.lower()
```

```python
df['employment'] = df['employment'].apply(make_lower)
```

```python
import random
def update_column(col_of_data):
    if random.randint(0,100) <= 5:
        print("---------------")
        print(col_of_data.name, col_of_data)
```

```python
# default is to apply the function down every column
df.apply(update_column)
```

```python
import random
def update_row(row):
    if random.randint(0,100) <= 1:
        print(row['age'], "|", row['gender'])
```

```python
df.apply(update_row, axis='columns')
```

### ApplyMap

Apply a function to every element in the dataframe

```python
def applymap_func(val):
    if random.randint(0,100) <= 1:
        print(type(val))
```

```python
df.applymap(applymap_func)
```

### Map

Series function that will use a dictionary to map from one value to another.

Any value not in the map will be changed to NAN

### Replace

Series function that will use a dictionary to map from one to another but it will leave unspecified values alone
